```python
import cv2 as cv
extract_fps = 10

folder_name, file_name = list(video_dict.items())[0]  ## manually select 0,1,2..., which stands for the video_dict arg
video_path = os.path.join(data_dir, file_name)
folder_dir=os.path.join(data_dir, folder_name)
if not os.path.exists(folder_dir): os.mkdir(folder_dir)
    
video_capturer = cv.VideoCapture(video_path)
fps_video = video_capturer.get(cv.CAP_PROP_FPS)
frame_width = int(video_capturer.get(cv.CAP_PROP_FRAME_WIDTH))
frame_height = int(video_capturer.get(cv.CAP_PROP_FRAME_HEIGHT))
print(f'Orginal video: fps: {fps_video}, size: {frame_width} x {frame_height}.')

cnt=0
is_opened = True
while(video_capturer.isOpened):
    is_opened, frame = video_capturer.read()
    if not is_opened: break
    
    cv.imshow(file_name,  frame)
    key = cv.waitKey(2) & 0xff
    if key == 27:  # escape key
        break
    if cnt%10 ==0: ## write image every 10 frames
        image_name = folder_name+'_'+str(cnt)+'.jpg'
        cv.imwrite(os.path.join(folder_dir, image_name), frame)
    cnt+=1
video_capturer.release()
cv.destroyAllWindows()
print('Done:', cnt, 'frames.')
```

## 2.  data load index/label generation

```python
'''
input: image folder gnerated above (with empty/nonempty subfolders)
output: csv file with image path and label
'''
import os
import re
import pandas as pd

root= os.environ['HOME']
data_dir = os.path.join(root,'./Documents/datasets/traffic')
folder_list = ['video2', 'video2_night']
data = pd.DataFrame(columns=['num','path','nonempty_road'])

for folder in folder_list:
    
    sub_folder = os.path.join(data_dir, folder) ## , e.g. './Documents/datasets/traffic/video2'
    
    ## add empty_road data
    label =0
    for file_name in os.listdir(os.path.join(sub_folder,'./empty_road')):
        num = cnt+int(re.findall(r'._([0-9]*).jpg',file_name)[0])  ## add sequence for sorting
        path = os.path.join(sub_folder,'./empty_road',file_name)
        data.loc[len(data)]=[num, path, label]

    ## add nonempty_road data
    label =1
    for file_name in os.listdir(os.path.join(sub_folder,'./nonempty_road')):
        num = cnt+int(re.findall(r'._([0-9]*).jpg',file_name)[0]) ## for distinguish video folder
        path = os.path.join(sub_folder,'./nonempty_road',file_name)
        data.loc[len(data)]=[num, path, label]

                      
data_sorted = data.sort_values(by = ['num'], axis=0, ascending= True)  ## axis=0
data_sorted.to_csv('../result/dataset_erd_pathLabel.csv', index = None)
print('Done!')
```
